```python
%matplotlib inline
import xlrd
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import os as os
from itertools import product
import scipy.stats as ss
from scipy.stats import norm
from math import e
np.set_printoptions(suppress=True, linewidth=200, edgeitems=100)
```

## Load Google Usage data, rename columns

```python
google_df_original = pd.read_csv("Google Storage Usage (10-JAN-2022).csv")
google_df_original              
```

```python
google_df_original.rename(columns={"User account status [2022-01-08 GMT]": "Account Status", 
                          "Admin status [2022-01-08 GMT]": "Admin Status",
                          "Gmail storage used (MB) [2022-01-08 GMT]": "Gmail Storage Used",
                          "Drive storage used (MB) [2022-01-08 GMT]": 'Drive Storage Used',
                          "Photos storage used (MB) [2022-01-08 GMT]": "Photo Storage Used",
                          "Total storage used (MB) [2022-01-08 GMT]": "Total Storage Used",
                          "Storage used (%) [2022-01-08 GMT]": "Storage Used Percentage"},inplace=True)
google_df_original
```

## Delete rows with no storage at all

```python
google_df = google_df_original[(google_df_original['Gmail Storage Used'] > 0) |
                               (google_df_original['Drive Storage Used'] > 0) |
                               (google_df_original['Photo Storage Used'] > 0) |
                               (google_df_original['Total Storage Used'] > 0)]
google_df
```

## Check if the last column has non-zero value, then drop

```python
google_df[google_df['Storage Used Percentage'] != 0]
```

```python
google_df_modified = google_df.drop(['Storage Used Percentage'],axis='columns')
google_df_modified
```

## Creat ID, fix user email to lower-case

```python
google_df_modified.insert(0,'User ID',range(1,1+len(google_df_modified)))
google_df_modified['User Email'] = google_df_modified['User'].str.lower()
google_df_modified
```

## Load and merge employee data with fixed email

```python
employee_data = pd.read_csv("Employee Emails for Merging.csv")
employee_data['Email'] = employee_data['EMAIL'].str.lower()
t1 = google_df_modified.merge(employee_data,how='left',left_on='User Email',
                              right_on='Email').drop_duplicates(subset=['User ID'])
t2 = t1.drop(['PERSON_ID','FULL_NAME','USERNAME','EMAIL','User','Email'],axis='columns')
t2['DEPARTMENT'] = t2['DEPARTMENT'].replace(np.nan,'Student and Special Accounts')
t2['PERSON_TYPE'] = t2['PERSON_TYPE'].replace(np.nan,'Student and Special Accounts')
t2
```

## Load and merge student data with fixed email

```python
student_data = pd.read_csv("Student Emails for Merging.csv")
student_data['Email'] = student_data['EMAIL'].str.lower()
t3 = t2.merge(student_data,how='left',left_on='User Email',right_on='Email').drop_duplicates(subset=['User ID'])
t4 = t3.drop(['PERSON_ID','USERNAME','EMAIL','ENTRANCE_YEAR','Email'],axis='columns')
t4
```

## Create User Type column

```python
t4['User Type'] = (t4['EMPLOYEE_TYPE'].astype(str) + t4['STUDENT_TYPE'].astype(str)).str.replace("nan","")
t4['User Type'] = [x if x != "" else "" for x in t4['User Type']]
t4['User Type'] = ['Special' if x == "" else x for x in t4['User Type']]
t4['User Type'] = ['Staff (Past and Present)' if x == 'Staff (Past and Present)Former Student' else x for x in t4['User Type']]
t4['User Type'] = ['Staff (Past and Present)' if x == 'Staff (Past and Present)Current Student' else x for x in t4['User Type']]
t4['User Type'] = ['Faculty (Past and Present)' if x == 'Faculty (Past and Present)Former Student' else x for x in t4['User Type']]
t4['User Type'] = ['Faculty (Past and Present)' if x == 'Faculty (Past and Present)Current Student' else x for x in t4['User Type']]
normal_account_df = t4[t4['User Type'] != 'Special']
normal_account_df
```

## Further categorize User Type = "Special"

```python
special_account_df = t4[t4['User Type'] == 'Special']
special_account_df
```

```python
special_df = special_account_df.drop(['STUDENT_TYPE','User Type'],axis='columns')
special_df['STUDENT_TYPE'] = ['Current Student' if ("_2026" in x) | ("_2025" in x) | ("_2024" in x) |
                                     ("_2023" in x) | ("_2022" in x) 
                                      else 'Former Student' if ("_2021" in x) | ("_2020" in x) |
                                     ("_2019" in x) | ("_2018" in x) | ("_2017" in x) | ("_2016" in x) |
                                     ("_2015" in x) | ("_2014" in x) | ("_2013" in x) | ("_2012" in x) |
                                     ("_2011" in x) | ("_2010" in x) else "" for x in special_account_df['User Email']]
special_df
```

```python
special_df['User Type'] = (special_df['EMPLOYEE_TYPE'].astype(str) + 
                           special_df['STUDENT_TYPE'].astype(str)).str.replace("nan","")
special_df['User Type'] = [x if x != "" else "" for x in special_df['User Type']]
special_df['User Type'] = ['Special' if x == "" else x for x in special_df['User Type']]
special_df
```

```python
google_df_clean = pd.concat([normal_account_df,special_df])
google_df_clean1 = google_df_clean[['User ID','Account Status','Admin Status','Gmail Storage Used',
                                  'Drive Storage Used','Photo Storage Used','Total Storage Used',
                                  'User Email','PERSON_TYPE','TITLE','DEPARTMENT','EMPLOYEE_TYPE',
                                  'STUDENT_TYPE','User Type']]
google_df_ready = google_df_clean1.rename(columns={'PERSON_TYPE':'Person Type',
                               'TITLE':'Title',
                               'DEPARTMENT':'Department',
                               'EMPLOYEE_TYPE':'Employee Type',
                               'STUDENT_TYPE':'Student Type'})
google_df_ready
```

```python
google_df_ready.to_csv("Google Usage Visualization.csv")
```
